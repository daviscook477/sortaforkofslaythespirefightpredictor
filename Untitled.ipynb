```python
import os
import glob
import json
import random
from multiprocessing import Pool
from tqdm import tqdm

data_path = r'C:\Users\davis\Documents\Slay the Spire Runs\2020-12-13'


def process_file(file_path):
    file_datas = []
    with open(file_path) as f:
        runs = json.load(f)
        for run in runs:
            if len(run) > 0:
                file_datas.append(random.choice(run))
    return file_datas


print('Looking for json data files')
file_paths = glob.glob(os.path.join(data_path, '*.json'))
print(f'Found {len(file_paths)} data files')

datas = []
for file_path in tqdm(file_paths):
    datas.extend(process_file(file_path))

print(len(datas))
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
import re
import numpy as np


def encode_list(list_to_encode, category):
    """
  Encodes a list of strings using the OneHotEncoder for a certain category.
  The number of times each string appears in the list being encoded is the value assigned to that string in the One-Hot Encoded vector.
  :param list_to_encode: list of strings to encode
  :param category: list of strings in the category in order
  """
    if len(list_to_encode) == 0:
        return np.zeros(len(category))
    np_array = np.array(list_to_encode)
    encoder = OneHotEncoder(categories=[category], sparse=False)
    n_by_1 = np_array.reshape(len(np_array), 1)
    onehot_encoded = encoder.fit_transform(n_by_1)
    summed = np.sum(onehot_encoded, axis=0)
    return summed


def encode_single(value, category):
    np_array = np.array([[value]])
    encoder = OneHotEncoder(categories=[category], sparse=False)
    onehot_encoded = encoder.fit_transform(np_array)
    collapsed = np.sum(onehot_encoded, axis=0)
    return collapsed


def generalize_strikes_and_defends(cards):
    """
  Modifies any character specific Strikes and Defends (e.g. Strike_R and Defend_R) into general Strikes and Defends (e.g. Strike and Defend).
  :param cards: the cards in the player deck that might need to be modified
  """
    for i, s in enumerate(cards):
        if s.startswith('Strike_') or s.startswith('Defend_'):
            cards[i] = re.sub('_.', '', s)


def unupgrade_cards(cards):
    for i, s in enumerate(cards):
        if s[-2] == '+':
            cards[i] = re.sub('\+.', '', s)


def encode_cards(cards):
    """
  Encodes a list of cards into a modified one-hot vector where each index represents how many of that card are in the deck.
  The vector has length of ALL_CARDS.
  :param cards: the cards in the player deck to be encoded into a vector
  """
    cards = np.array(cards)
    unupgrade_cards(cards)
    generalize_strikes_and_defends(cards)
    filter_arr = [(card in ALL_IRONCLAD_CARDS) for card in cards]
    cards = cards[filter_arr]
    return encode_list(cards, ALL_IRONCLAD_CARDS)


def encode_relics(relics):
    """
  Encodes a list of relics into a modified one-hot vector of length ALL_RELICS.
  If the relic is present in relics, it will be represented as 1 in the returned
  vector
  """
    relics = np.array(relics)
    filter_arr = [(relic in ALL_IRONCLAD_RELICS) for relic in relics]
    relics = relics[filter_arr]
    return encode_list(relics, ALL_IRONCLAD_RELICS)


def encode_sample_with_loop(sample):
    """
  Encode a single sample into a 1D input vector for the model.
  :param sample: the training sample as json data (i.e. nested dictionary/list structure in python)
  """
    deck = encode_cards(sample['deck'])
    relics = encode_relics(sample['relics'])

    noptions = np.array(sample['not_picked'])
    if sample['picked'] != 'SKIP' and sample['picked'] != 'Singing Bowl':
        noptions = np.append(noptions, [sample['picked']])

    options = encode_cards(noptions)
    options = np.append(options, [1.0])

    # return (np.append(deck, [sample['floor']]), options)
    return (np.append(deck, [sample['floor']]), relics, options)
```

```python
import random


def preprocess_with_loop(data, log=True, log_freq=1000):
    """
  Pre-processes the data one sample at a time using loops.
  X has shape: (num_samples, training_vec_len)
  Y has shape: (num_samples)
  """
    processed_samples = []
    processed_samples_r = []
    processed_aux = []
    y = []

    if log:
        print(f'Processing {len(data)} samples')

    for i, sample in enumerate(data):
        # logging
        if log and i % log_freq == 0:
            print(
                f'{((i / len(data)) * 100):.1f} % complete. => {i} of {len(data)}'
            )

        # only ironclad
        if sample['character'] != 'IRONCLAD':
            continue

        # no old relics
        if 'Dodecahedron' in sample['relics']:
            continue
            
        # actual processing into expected data format
        noptions = np.array(sample['not_picked'])
        if sample['picked'] != 'SKIP' and sample['picked'] != 'Singing Bowl':
            np.append(noptions, [sample['picked']])
        unupgrade_cards(noptions)
        failure = False
        for opt in noptions:
            if not opt in ALL_IRONCLAD_CARDS:
                failure = True
                break
        if failure or len(noptions) <= 1:
            continue

        if sample['picked'] != 'SKIP' and sample['picked'] != 'Singing Bowl':
            unupgraded_pick = sample['picked']
            if unupgraded_pick[-2] == '+':
                unupgraded_pick = re.sub('\+.', '', unupgraded_pick)
            if not unupgraded_pick in ALL_IRONCLAD_CARDS:
                continue
            encoded_picked = encode_cards([sample['picked']])
            encoded_picked = np.append(encoded_picked, [0.0])
        else:
            encoded_picked = np.zeros(len(ALL_IRONCLAD_CARDS) + 1)
            encoded_picked[len(encoded_picked) - 1] = 1.0

        (p_sample, p_sample_r, p_aux) = encode_sample_with_loop(sample)
        processed_samples.append(p_sample)
        processed_samples_r.append(p_sample_r)
        processed_aux.append(p_aux)
        y.append(encoded_picked)

    X1 = np.vstack(processed_samples)
    R = np.vstack(processed_samples_r)
    X2 = np.vstack(processed_aux)
    Y = np.vstack(y)
    return X1, R, X2, Y
```

```python
def scale_X(X_data):
    """
  Used with one hot encoded model
  """
    X_copy = np.copy(X_data)
    for sample in X_copy:
        deck_size = 0
        for i in range(len(ALL_IRONCLAD_CARDS)):
            deck_size += sample[i]
        for i in range(len(ALL_IRONCLAD_CARDS)):
            sample[i] /= deck_size
    return X_copy


X1_scaled = scale_X(X1)
R_scaled = R
X2_scaled = X2
Y_scaled = Y
```

```python
import tensorflow.keras as keras
import tensorflow as tf
import os
import datetime
import tensorflow.keras.backend as K
from sklearn.model_selection import train_test_split
import sys

X1_train, X1_test, R_train, R_test, X2_train, X2_test, Y_train, Y_test = train_test_split(
    X1_scaled, R_scaled, X2_scaled, Y_scaled, test_size=0.2, shuffle=True)

train_size = len(Y_train)
test_size = len(Y_test)

batch_size = 1024

train_dataset = tf.data.Dataset.from_tensor_slices((X1_train, R_train, X2_train, Y_train))
train_dataset = train_dataset.batch(batch_size)
test_dataset = tf.data.Dataset.from_tensor_slices((X1_test, R_test, X2_test, Y_test))
test_dataset = test_dataset.batch(batch_size)

def avg_lambda_fun(x, mask):
    mask_cast = K.cast(mask, 'float64')
    expanded = K.expand_dims(mask_cast)
    count = K.sum(mask_cast)
    sum = K.sum(expanded * x, axis=1)
    return sum / count

## two input layers to the model, pass both to the loss function
# input = tf.keras.layers.Input(75 + 1, dtype=tf.float64)
input = tf.keras.layers.Input(75 + 1, dtype=tf.float64)

# relic_input = tf.keras.layers.Input(shape=(150, ), dtype=tf.float64)
# relic_embedding = tf.keras.layers.Embedding(150, 5, mask_zero=True, dtype=tf.float64)(relic_input)
# relic_average = tf.keras.layers.Lambda(avg_lambda_fun, output_shape=(15, ), mask=None, dtype=tf.float64)(relic_embedding)

# merged_input = tf.keras.layers.Concatenate(axis=1, dtype=tf.float64)([input, relic_average])

hidden = tf.keras.layers.Dense(100, activation='relu', dtype=tf.float64)(input)
drop = tf.keras.layers.Dropout(0, dtype=tf.float64)(hidden)
output = tf.keras.layers.Dense(75 + 1, dtype=tf.float64)(drop)
model = tf.keras.models.Model(inputs=input, outputs=output)


optimizer = tf.keras.optimizers.Adam(learning_rate=0.005)

# random draft
# def random_draft_error(x1, x2, y):
#     r = np.random.rand(x2.shape[0], 75 + 1)
#     mr = tf.math.multiply(x2, r)
#     idx_y = K.argmax(y, axis=1)
#     idx_mr = K.argmax(mr, axis=1)
#     incorrect = K.not_equal(idx_y, idx_mr)
#     error = K.cast(incorrect, tf.float64)
#     return error

# rare draft
# def rare_draft_error(x1, x2, y):
    # multiply the options by a rare mask
    # if there is a non-zero entry, pick
    # randomly and then compare with the
    # real choice
    # this probably cannot be done as a tensor
    # operation and will require me to process
    # row by row

train_loss_results = []
train_error_results = []
num_epochs = 50

for epoch in range(num_epochs):
    epoch_loss_avg = tf.keras.metrics.Mean()
    epoch_failure_count = tf.keras.metrics.Sum()

    for x1, r, x2, y in train_dataset:
        with tf.GradientTape() as tape:
            y_ = model(x1, training=True)
            loss_value = K.sum(K.square(tf.math.subtract(y, tf.math.multiply(x2, y_))))
        grads = tape.gradient(loss_value, model.trainable_variables)
        optimizer.apply_gradients(zip(grads, model.trainable_variables))
        epoch_loss_avg.update_state(loss_value)
        
        my_ = tf.math.multiply(x2, y_)
        idx_y = K.argmax(y, axis=1)
        idx_my_ = K.argmax(my_, axis=1)
        incorrect = K.not_equal(idx_y, idx_my_)
        error = K.cast(incorrect, tf.float64)
        epoch_failure_count.update_state(error)

    train_loss_results.append(epoch_loss_avg.result())
    train_error_results.append(epoch_failure_count.result())
    
    epoch_loss_avg_test = tf.keras.metrics.Mean()
    epoch_failure_count_test = tf.keras.metrics.Sum()
    
    for x1_t, r_t, x2_t, y_t in test_dataset:
        y__t = model(x1_t, training=False)
        loss_value = K.sum(K.square(tf.math.subtract(y_t, tf.math.multiply(x2_t, y__t))))
        epoch_loss_avg_test.update_state(loss_value)
        
        my__t = tf.math.multiply(x2_t, y__t)
        idx_y_t = K.argmax(y_t, axis=1)
        idx_my__t = K.argmax(my__t, axis=1)
        incorrect = K.not_equal(idx_y_t, idx_my__t)
        error = K.cast(incorrect, tf.float64)
        epoch_failure_count_test.update_state(error)
        
#     epoch_random_failure_count_test = tf.keras.metrics.Sum()
    
#     for x1_t, x2_t, y_t in test_dataset:
#         error = random_draft_error(x1_t, x2_t, y_t)
#         epoch_random_failure_count_test.update_state(error)

    # , random test failure {epoch_random_failure_count_test.result().numpy() / test_size}
    print(f'epoch {epoch}, train loss {epoch_loss_avg.result().numpy()}, train failure {epoch_failure_count.result().numpy() / train_size}, test loss {epoch_loss_avg_test.result().numpy()}, test failure {epoch_failure_count_test.result().numpy() / test_size}')
```
